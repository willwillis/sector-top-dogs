```python
# Initial imports
import pandas as pd
import hvplot.pandas
from sqlalchemy import create_engine
import pickle
```

```python
# create a connection to the database using the create_engine function
engine = create_engine("postgresql://postgres:2242020@localhost:5432/russell3000stockdata")
```

```python
# Define a query that select all rows from the animals_all table
query = "SELECT * FROM ticker_data"

# Load data into the DataFrame using the read_sql() method from pandas
tickers_df = pd.read_sql(query, engine)
```

```python
ticker = "AAPL"
start = '2006-01-01'
end = '2018-12-31'
```

```python
pickle_file = f"../../data/rawdata/stock-data-by-range/{start}_{end}/{ticker}.pickle"
```

```python
df = pickle.load( open( pickle_file, "rb" ) )
```

```python
df.reset_index(inplace = True)
df['ticker'] = ticker
df.drop(columns=['Adj Close'], inplace= True)
df['year'] = df['Date'].dt.year
df.rename(columns={'Date':'trading_day', 'High':'high', 'Low': 'low', 'Open': 'open', 'Close':'close', 'Volume':'volume'}, inplace=True)
df.to_sql('ticker_data', con=engine, if_exists='append')
```

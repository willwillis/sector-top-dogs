# Parse a CSV

```python
import numpy as np
import pandas as pd
import os.path
import csv
from pathlib import Path
# again, overkill on the modules
```

## Read composition file

```python

russell = Path('../../data/rawdata/fund-composition-as-of/20061031.csv')

all_tickers = []

with open(russell, 'r') as holdings:
    csvreader = csv.reader(holdings, delimiter=',')
    etf_name = next(csvreader)
    inception_date = next(csvreader)
    holdings_as_of_date = next(csvreader)
    net_assets = next(csvreader)
    
    for junk in range(7):
        next(csvreader)
        
    for row in csvreader:
        ticker = row[0]
        asset_class = row[2]
        if asset_class == 'Equity':
            print(f"{ticker} is an {asset_class}")
            all_tickers.append(ticker)
```
